```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    step = 1

    print(f"\nПочинаємо пошук у відсортованому масиві: {arr}")

    while low <= high:
        # Обчислюємо індекс середини
        mid = (low + high) // 2
        guess = arr[mid]

        print(f"Крок {step}: Перевіряємо елемент на індексі {mid} (значення: {guess})")

        if guess == target:
            print(f"✅ Успіх! Число {target} знайдено на позиції {mid}.")
            return True

        if guess > target:
            # Число замале, шукаємо в лівій половині
            high = mid - 1
        else:
            # Число завелике, шукаємо в правій половині
            low = mid + 1

        step += 1

    print(f"❌ Число {target} відсутнє у списку.")
    return False

def main():
    try:
        # Введення даних
        raw_input = input("Введіть список чисел через кому: ")
        # Перетворення у список чисел
        numbers = [int(x.strip()) for x in raw_input.split(",")]

        target = int(input("Введіть число для пошуку: "))

        # Бінарний пошук вимагає ВІДСОРТОВАНОГО масиву
        numbers.sort()

        binary_search(numbers, target)

    except ValueError:
        print("Помилка: введіть коректні цілі числа через кому.")

if __name__ == "__main__":
    main()
```
